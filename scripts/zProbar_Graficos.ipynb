```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Generar datos simulados de ventas (dfv)
fechas = pd.date_range(end=datetime.today(), periods=400)
dfv = pd.DataFrame({
    "Fecha": fechas,
    "Codigo_Articulo": [123] * len(fechas),
    "Sucursal": [1] * len(fechas),
    "Unidades": np.random.poisson(lam=10, size=len(fechas))
})

# Generar datos simulados de stock (dfs)
dfs = pd.DataFrame({
    "c_articulo": [123],
    "c_sucu_empr": [1],
    "c_anio": [datetime.today().year],
    "c_mes": [datetime.today().month],
    **{f"q_dia{d}": [np.random.randint(0, 100)] for d in range(1, 29)}
})

# Generar datos simulados de ofertas (dfo)
dfo = pd.DataFrame({
    "c_articulo": [123],
    "c_sucu_empr": [1],
    "c_anio": [datetime.today().year],
    "c_mes": [datetime.today().month],
    **{f"m_oferta_dia{d}": [np.random.choice([0, 1])] for d in range(1, 29)}
})

# Valores ejemplo para los parámetros de forecast y ventas históricas
forecast = 300.0
average = 10.5
ventas_last = 310.0
ventas_previous = 280.0
ventas_same_year = 260.0

# Ejecutar la función
resultado = generar_grafico_json(
    dfv=dfv,
    dfs=dfs,
    dfo=dfo,
    articulo=123,
    sucursal=1,
    Forecast=forecast,
    Average=average,
    ventas_last=ventas_last,
    ventas_previous=ventas_previous,
    ventas_same_year=ventas_same_year,
    convertir_stock_diario_a_dict=convertir_stock_diario_a_dict,
    convertir_ofertas_a_dict=convertir_ofertas_a_dict
)

# Mostrar resultados
import json
print(json.dumps(resultado, indent=2, ensure_ascii=False))
```
